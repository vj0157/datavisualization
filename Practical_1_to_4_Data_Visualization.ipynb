

# handle Outlier

```python
# Import required libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

# Reading the data
df = pd.read_csv("real_estate.csv")
print(df.shape)
print(df.info())
```

```python
df.describe()
```

```python
# Identifying Outliers with Interquartile Range (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print(df < (Q1 - 1.5 * IQR)) == (df > (Q3 + 1.5 * IQR));
```

```python
# Identifying Outliers with Skewness
print(df['ST_NUM'].skew())
df['ST_NUM'].describe()
```

```python
#Identifying Outliers with Visualization
# 1. Box Plot
plt.boxplot(df["ST_NUM"])
plt.show()
# 2. Histogram
df.ST_NUM.hist()
# 3. Scatterplot
```

```python
# Replacing Outliers with Median Values
print(df['ST_NUM'].quantile(0.50)) 
print(df['ST_NUM'].quantile(0.95)) 
df['ST_NUM'] = np.where(df['ST_NUM'] > 325, 140, df['ST_NUM'])
df.describe()
```

4 pract :Perform Reshaping of the hierarchical data and pivoting data frame data.

```python
# import pandas module
import pandas as pd

# making dataframe
df = pd.read_csv("nba.csv")


# it was print the first 5-rows
print(df.head())
```

```python
# reshape the dataframe using stack() method
df_stacked = df.stack()
  
print(df_stacked.head(26))
  
```

```python
# reshape the dataframe using unstack() method
df_unstacked = df_stacked.unstack()
print(df_unstacked.head(10))
```

```python
# reshape the dataframe using melt() method
# it takes two columns "Name" and "Team"
df_melt = df.melt(id_vars =['Name', 'Team']) 
print(df_melt.head(10))
```

```python
import numpy as np
import pandas as pd
```

```python
# making dataframe
df = pd.DataFrame({'fff': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bbb': ['P', 'Q', 'R', 'P', 'Q', 'R'],
                   'baa': [2, 3, 4, 5, 6, 7],
                   'zzz': ['h', 'i', 'j', 'k', 'l', 'm']})
df
```
